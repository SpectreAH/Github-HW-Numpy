Place 10 points randomly on a plane, then calculate the distance vectors between these different points. How many distance calculations do you do? Visualize points with matplotlib

```python
#10 random points on a plane
points = np.random.rand(10, 2)

#The distance vectors between different points
dist = []
for i in range(len(points)):
    for j in range(i + 1, len(points)):
        distvect = points[j] - points[i]
        dist.append(np.linalg.norm(distvect))

#Total number of distance calculations
numdiscal = len(dist)

#To visualise with matplotlib
plt.scatter(points[:, 0], points[:, 1], label='Points')
plt.title('Randomly Generated Points on a Plane')
plt.xlabel('X axis')
plt.ylabel('Y axis')
plt.legend()
plt.show()

#Print statements
print("The distances between points is")
print(dist)
print("The number of distance calculations is", numdiscal)
```
